```python
import pandas as pd

air_data = pd.read_csv('Project1_Data/AirQuality_Daily_StudentVersion(in).csv')
air_data = pd.DataFrame(air_data)
```

```python
def assign_temperature(temperature):
    if temperature < 32:
        return 'Below freezing'
    elif 32 <= temperature < 50:
        return 'Cool'
    elif 50 <= temperature < 70:
        return 'Warm'
    else:
        return 'Hot'

air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature)
```

```python
def assign_altitude(altitude):
    if altitude < 2000:
        return 'Low'
    elif 2000 <= altitude <= 3000:
        return 'Medium'
    else:
        return 'High'

air_data['altitude_level'] = air_data['sensor.altitude'].apply(assign_altitude)
```

```python
def assign_humidity(humidity):
    if humidity < 50:
        return 'Low'
    elif 50 <= humidity < 80:
        return 'High'
    else:
        return 'Very high'

air_data['humidity_level'] = air_data['humidity'].apply(assign_humidity)
```

```python
air_data_grouped = air_data.groupby(['sensor.name'])


air_data_aggregated_pm2_5 = air_data_grouped.agg(median_pm2_5 = ('pm2.5_atm', 'median'),
                                                  mean_pm2_5 = ('pm2.5_atm', 'mean'))
top_5_pm2_5 = air_data_aggregated_pm2_5.sort_values(by='mean_pm2_5', ascending=False).head(5)


air_data_aggregated_pm10_0 = air_data_grouped.agg(median_pm10_0 = ('pm10.0_atm', 'median'),
                                                  mean_pm10_0 = ('pm10.0_atm', 'mean'))
top_5_pm10_0 = air_data_aggregated_pm10_0.sort_values(by='mean_pm10_0', ascending=False).head(5)


air_data_aggregated_voc = air_data_grouped.agg(median_voc = ('voc', 'median'),
                                                mean_voc = ('voc', 'mean'))
top_5_voc = air_data_aggregated_voc.sort_values(by='mean_voc', ascending=False).head(5)


print(top_5_pm2_5)
print(top_5_pm10_0)
print(top_5_voc)

```
